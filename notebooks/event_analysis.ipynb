```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from datetime import datetime

# Connect to the database
engine = create_engine('postgresql://postgres:POSTgres123%40@localhost/ems_db')

# Query the data
events_df = pd.read_sql_query('SELECT * FROM events', con=engine)
attendees_df = pd.read_sql_query('SELECT * FROM attendees', con=engine)
users_df = pd.read_sql_query('SELECT * FROM users', con=engine)

# Analysis and visualization

# 1. Number of attendees per event
attendees_per_event = attendees_df.groupby('event_id').size()
attendees_per_event.plot(kind='bar')
plt.title('Number of Attendees per Event')
plt.xlabel('Event ID')
plt.ylabel('Number of Attendees')
plt.show()

# 2. User registration trends over time
users_df['created_at'] = pd.to_datetime(users_df['created_at'])
user_registration_trends = users_df.set_index('created_at').resample('M').size()
user_registration_trends.plot(kind='line')
plt.title('User Registration Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Registrations')
plt.show()

# 3. Event popularity (number of attendees per event)
event_popularity = attendees_df.groupby('event_id').size().reset_index(name='num_attendees')
event_popularity = pd.merge(event_popularity, events_df, on='event_id')
event_popularity = event_popularity[['title', 'num_attendees']]
event_popularity = event_popularity.sort_values(by='num_attendees', ascending=False)
event_popularity.plot(kind='bar', x='title', y='num_attendees')
plt.title('Event Popularity')
plt.xlabel('Event Title')
plt.ylabel('Number of Attendees')
plt.show()

# 4. Distribution of event dates
events_df['event_date'] = pd.to_datetime(events_df['event_date'])
event_dates_distribution = events_df['event_date'].dt.date.value_counts().sort_index()
event_dates_distribution.plot(kind='bar')
plt.title('Distribution of Event Dates')
plt.xlabel('Event Date')
plt.ylabel('Number of Events')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

# Connect to the database
engine = create_engine('postgresql://postgres:POSTgres123%40@localhost/ems_db')

# Query the data
events_df = pd.read_sql_query('SELECT * FROM events', con=engine)
attendees_df = pd.read_sql_query('SELECT * FROM attendees', con=engine)
users_df = pd.read_sql_query('SELECT * FROM users', con=engine)
venues_df = pd.read_sql_query('SELECT * FROM venues', con=engine)

# Analyze and visualize data

# 1. Number of attendees per event
attendees_per_event = attendees_df.groupby('event_id').size()
print("Number of Attendees per Event:")
print(attendees_per_event)
attendees_per_event.plot(kind='bar')
plt.title('Number of Attendees per Event')
plt.xlabel('Event ID')
plt.ylabel('Number of Attendees')
plt.show()

# Pie chart for number of attendees per event
print("Percentage of Attendees per Event:")
print(attendees_per_event / attendees_per_event.sum() * 100)
attendees_per_event.plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Attendees per Event')
plt.ylabel('')
plt.show()

# 2. User registration trends over time
users_df['created_at'] = pd.to_datetime(users_df['created_at'])
user_registration_trends = users_df.set_index('created_at').resample('M').size()
print("User Registration Trends Over Time:")
print(user_registration_trends)
user_registration_trends.plot(kind='line')
plt.title('User Registration Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Registrations')
plt.show()

# Pie chart for user registrations by month
print("Percentage of User Registrations by Month:")
print(user_registration_trends / user_registration_trends.sum() * 100)
user_registration_trends.plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of User Registrations by Month')
plt.ylabel('')
plt.show()

# 3. Event popularity (number of attendees per event)
event_popularity = attendees_df.groupby('event_id').size().reset_index(name='num_attendees')
event_popularity = pd.merge(event_popularity, events_df, on='event_id')
event_popularity = event_popularity[['title', 'num_attendees']]
event_popularity = event_popularity.sort_values(by='num_attendees', ascending=False)
print("Event Popularity (Number of Attendees per Event):")
print(event_popularity)
event_popularity.plot(kind='bar', x='title', y='num_attendees')
plt.title('Event Popularity')
plt.xlabel('Event Title')
plt.ylabel('Number of Attendees')
plt.show()

# Pie chart for event popularity
print("Percentage of Event Popularity:")
print(event_popularity.set_index('title')['num_attendees'] / event_popularity['num_attendees'].sum() * 100)
event_popularity.set_index('title').plot(kind='pie', y='num_attendees', autopct='%1.1f%%')
plt.title('Percentage of Event Popularity')
plt.ylabel('')
plt.show()

# 4. Distribution of event dates
events_df['event_date'] = pd.to_datetime(events_df['event_date'])
event_dates_distribution = events_df['event_date'].dt.date.value_counts().sort_index()
print("Distribution of Event Dates:")
print(event_dates_distribution)
event_dates_distribution.plot(kind='bar')
plt.title('Distribution of Event Dates')
plt.xlabel('Event Date')
plt.ylabel('Number of Events')
plt.show()

# Pie chart for distribution of event dates
print("Percentage of Event Dates:")
print(event_dates_distribution / event_dates_distribution.sum() * 100)
event_dates_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Event Dates')
plt.ylabel('')
plt.show()

# 5. Number of events per user
events_per_user = events_df.groupby('user_id').size()
print("Number of Events per User:")
print(events_per_user)
events_per_user.plot(kind='bar')
plt.title('Number of Events per User')
plt.xlabel('User ID')
plt.ylabel('Number of Events')
plt.show()

# Pie chart for number of events per user
print("Percentage of Events per User:")
print(events_per_user / events_per_user.sum() * 100)
events_per_user.plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Events per User')
plt.ylabel('')
plt.show()

# 6. Number of events per venue
events_per_venue = events_df.groupby('venue_id').size()
print("Number of Events per Venue:")
print(events_per_venue)
events_per_venue.plot(kind='bar')
plt.title('Number of Events per Venue')
plt.xlabel('Venue ID')
plt.ylabel('Number of Events')
plt.show()

# Pie chart for number of events per venue
print("Percentage of Events per Venue:")
print(events_per_venue / events_per_venue.sum() * 100)
events_per_venue.plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Events per Venue')
plt.ylabel('')
plt.show()

# 7. Average number of attendees per event
average_attendees = attendees_per_event.mean()
print(f'Average number of attendees per event: {average_attendees}')
```
